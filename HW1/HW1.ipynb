```python
# 引入套件
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rcParams['font.family'] = 'Arial Unicode MS'
```

```python
# 設定 CSV 檔案(stock_index)的路徑
csv_file_path = 'JUNE_FIRE.csv'

# 指定要跳過的行數，：
skip_rows = [1]

# 使用 pandas 的 read_csv 函式讀取 CSV 檔案並將其轉換為 DataFrame 並使用 utf-8 的編碼方式來處理這個檔案，並跳過第一行
df = pd.read_csv(csv_file_path , encoding = 'utf-8' ,skiprows=skip_rows )


df
```

```python
# 儲存發生縣市的數據

FireCount = df ['受理火災件數']

FireCount
```

```python
# 設定圖表標題(發生區域與次數的關係)

# 提取區域名和火災次數的資料
districts = df['區域別']  # 替換'區域名稱'為您的CSV中區域名稱的欄位名稱
fire_counts = df['受理火災件數']  # 替換'火災次數'為您的CSV中火災次數的欄位名稱

# 創建一張條形圖
plt.bar(districts, fire_counts)

# 添加標題和軸標籤
plt.title('112年6月雲林縣不同區域發生火災之次數')
plt.xlabel('區域別')
plt.ylabel('次數')

# 設定x軸標籤垂直顯示以避免重疊
plt.xticks(rotation=65)

# 顯示圖表
plt.show()
```

```python
# 設定 CSV 檔案(stock_index)的路徑
csv_file_paths = ['JAN_FIRE.csv','FEB_FIRE.csv','MAR_FIRE.csv','APR_FIRE.csv','MAY_FIRE.csv','JUNE_FIRE.csv']
custom_x_labels = ['一月', '二月', '三月', '四月', '五月', '六月']  # 自訂的x軸標籤


# 創建一個空的資料框，稍後將合併所有檔案的內容
combined_data = pd.DataFrame()

# 逐一讀取並合併CSV檔案
for file_path in csv_file_paths:
    df = pd.read_csv(file_path , encoding = 'utf-8')
    
# 讀取每個檔案中的某一單元格，例如第3行（以0為基數的索引）和第5列（以0為基數的索引）
    data_in_cell = df.iloc[0, 1]  # 這裡的2表示第3行，4表示第5列
    
# 將讀取的數據添加到combined_data中
    combined_data = combined_data.append({'Data': data_in_cell}, ignore_index=True)
    
# 顯示合併後的資料
print("Combined Data:")
print(combined_data)
```

```python
# 繪製直條圖
plt.figure(figsize=(8, 6))
plt.bar(custom_x_labels, combined_data['Data'])
plt.xlabel('月份')
plt.ylabel('次數')
plt.title('雲林縣一月份到六月份之火災發生次數')


# 顯示直條圖
plt.show()
```

```python
# 設定 CSV 檔案的路徑
csv_file_path = 'JUNE_FIRE.csv'

# 使用 pandas 的 read_csv 函式讀取 CSV 檔案
df = pd.read_csv(csv_file_path, encoding='utf-8')

# 找到'總計'這一列的所有數據
total_row = df[df['區域別'] == '總計']

# 提取總計這一列的所有數據（不包括'區域別'列）
total_data = total_row.iloc[:, 1:]

# 顯示總計數據
print(total_data)

```

```python
# 提取所需的數據，不包括'出動人次-合計'
total_data = total_row[['出動消防人員', '出動義消人員', '出動其他人次']]

# 切片出需要的數據
labels = total_data.columns
sizes = total_data.values[0]
explode = (0.1, 0, 0)

# 繪製圓餅圖
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)
plt.axis('equal')  # 使圓餅圖比例相等
plt.title('出動人次分佈', y=1.05)

# 顯示圓餅圖
plt.show()
```
