```python
import fitz  # PyMuPDF
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import jieba

# 載入 PDF 文件
pdf_path = r"C:\Users\MIRDC\OneDrive\文件\GitHub\112-1PL\HW5\十二年國民基本教育課程綱要國民中學暨普通型高級中等學校─科技領域.pdf"
doc = fitz.open(pdf_path)

# 提取 PDF 文本
pdf_text = ""
for page_num in range(doc.page_count):
    page = doc[page_num]
    pdf_text += page.get_text()

# 關閉 PDF 文件
doc.close()

# 文本處理和分詞
seg_list = jieba.lcut(pdf_text)

# 移除停用詞
stopwords = ["的", "和", "是", "在", "之", "與", "等", "及"]
filtered_words = [word for word in seg_list if word not in stopwords]

# 計算詞頻
word_freq = Counter(filtered_words)

# 使用支援中文的字型
font_path = "C:\\Windows\\Fonts\\msyh.ttc"  # Microsoft YaHei

# 生成文字雲，指定字體路徑
wordcloud = WordCloud(width=800, height=400, background_color="white", font_path=font_path).generate(" ".join(filtered_words))

# 顯示文字雲
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']  # 請替換成你系統中的中文字型
plt.rcParams['axes.unicode_minus'] = False  # 解決負號顯示問題

# 新增要過濾的連接詞
stopwords.extend(["與", "之", "及", "IV", "V", "能", "人", "資", "生", "設", "性"])

# 移除非字母的詞和停用詞
filtered_words = [word for word in seg_list if word.isalpha() and word not in stopwords]

# 計算詞頻
word_freq = Counter(filtered_words)

# 顯示前 10 個最常見的詞及其詞頻
top_words = word_freq.most_common(10)
for word, freq in top_words:
    print(f"{word}: {freq} 次")

# 分離詞和詞頻
words, frequencies = zip(*top_words)

# 繪製直條圖
plt.bar(words, frequencies, color='skyblue')
plt.xlabel('詞彙')
plt.ylabel('詞頻')
plt.title('前 10 大詞條詞頻分析')
plt.show()
```
